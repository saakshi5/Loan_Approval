<font size="6">7.Splitting Data into x and y<font>

```python
# splitting x and y into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

<font size='6'>8.Importing Required Libraries for Model</font>

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve
```

```python
# creating instance of class
lr=LogisticRegression()
kn=KNeighborsClassifier()
svm=SVC()
dt=DecisionTreeClassifier()
```

Building Logistic Model

```python
def reg(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    print('classification report')
    print(classification_report(ytest,ypred))
    print('-'*50)
    print('Training accuracy',model.score(xtrain,ytrain))
    print('Testing accuracy',model.score(xtest,ytest))
    print('accuracy score',accuracy_score(ypred,ytest))
```

```python
# building logistic regression
reg(lr)
```

The classification report along with the high accuracy scores (both training and testing) indicate a very strong performance of the model on the loan approval dataset. Here are the insights we can gather from these results:

High Model Performance: The precision, recall, and f1-score for both classes (0 and 1) are perfect (1.00), indicating that the model predicts both classes accurately without any false positives or false negatives.
Perfect Accuracy: The accuracy score of 1.0 (100%) for both training and testing datasets suggests that the model is able to classify loan approval outcomes with complete accuracy.
No Overfitting or Underfitting: The identical accuracy scores for training and testing datasets (1.0) indicate that the model is well-generalized and not overfitting to the training data.

```python
#Predicting probabilities 
ypredprob=lr.predict_proba(xtest)[:,1]
```

```python
#Drawing roc_auc curve
fpr,tpr,thres=roc_curve(ytest,ypredprob)
plt.plot(fpr,tpr)
plt.show()
```

Thus,the ROC curve confirms the perfect performance indicated by the classification report, providing visual validation of the model's ability to accurately classify loan approval outcomes with no false positives or false negatives.

KNN Model

```python
#calculating best k values
ac_lst=[]
for i in range(1,21):
    kn=KNeighborsClassifier(n_neighbors=i)
    kn.fit(xtrain,ytrain)
    ypredd=kn.predict(xtest)
    ac=accuracy_score(ytest,ypredd)
    ac_lst.append(ac)
```

```python
#identifying best value of k by plotting
plt.plot(range(1,21),ac_lst,marker='o')
```

Here, best value of k is 11

```python
#using k=11
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(xtrain,ytrain)
ypredd=kn.predict(xtest)
print('classification report')
print(classification_report(ytest,ypredd))
print('-'*50)
print('Training accuracy',knn.score(xtrain,ytrain))
print('Testing accuracy',knn.score(xtest,ytest))
print('accuracy score',accuracy_score(ypredd,ytest))
```

Based on the updated classification report and model performance metrics KNN classifier with n_neighbors=11, here are the insights:

Precision and Recall:
The precision for class 1 (loan approved) has increased to 0.80, indicating that when the model predicts a loan approval, it is correct 80% of the time. The recall for class 1 is also high at 0.92, meaning the model correctly identifies 92% of actual loan approvals

F1-Score:
The F1-score for class 1 (loan approved) has improved significantly to 0.85, suggesting a better balance between precision and recall for this class. The F1-score for class 0 (loan not approved) remains relatively low at 0.26, indicating challenges in predicting this class accurately.

Accuracy:
The overall accuracy of the model is 0.76, representing the proportion of correctly predicted outcomes (loan approvals and rejections) out of the total predictions. While the accuracy has improved compared to the previous mode.

The training accuracy (0.699) is lower than the testing accuracy (0.774), which suggests a model that is adequately fitting the training data but may benefit from further optimization to enhance generalization.
